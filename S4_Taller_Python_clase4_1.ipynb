## Cargando Datos con Pandas

Pandas ofrece várias funciones para cargar datos de una diveridad de formatos.   
Por ejemplo la funcion <b>read_csv()</b>  sirve para leer data tabular de un archivo de texto separado por comas (tambien puede ser otro separador).

```python
#Nota: recordar que a veces no se lee la data completa en formato .xlsx.
  # Se prefiere mejor el formato .csv, sobre todo si los valores de algunas columnas son decimales

filename = 'https://github.com/luccemhu/Python-introduccion-Qlab/raw/main/S4-pima-indians-diabetes.data.csv' # Guardamos la data en un objeto
column_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class'] # Asignamos nombres a cada columna (tambien llamados atributos), ya que no lo tienen.

data = pd.read_csv(filename, names=column_names) # Importamos la data
data
# Class: tuvo diabete despues de 5 años<- 0: No ; 1: Sí
```

 - promedio: `.mean()`
 - desviacion standard: `.std()`

Por defecto las funciones de reduccion se hacen a nivel de columnas (axis=0). Para hacer la reduccion a nivel de filas debemos pasar el argumento <b>axis=1</b> a la operacion requerida

### Viendo si hay valores faltantes
La funcion <b>isnull()</b> de un DataFrame marca cada celda NaN con True.

```python
data.isnull() # Muestra las celdas con valores nulos en True
```

```python
data.isnull().values.any() # Muestra si hay valores Nan o no (No cuantos)
```

### Eliminando filas o columnas con valores faltantes

La funcion <b>dropna()</b> devuelve un DataFrame sin las filas (o columnas si se especifica argumento axis=1)

```python
data.shape # Filas y columnas
```

```python
data1 = data.dropna() # Subseteamos, eliminando filas con valores NaN
data1.head(10) # Los 10 primeros:
```

```python
data1.shape # Ahora las filas y columnas
```

```python
data1.isnull().sum()  # Muestra las columnas con la cantidad de filas con valores nulos:
```

```python
data.dropna(inplace = True) # Se eliminen los NanN en la misma data y no se cree otra.
data.head()
```

### Imputando valores faltantes

La funcion <b>fillna()</b> devuelve un DataFrame con valores faltantes imputados a nivel de columnas. Se le puede pasar como valores a imputar las medias de las columnas

```python
data.mean() # Media de la data
```

```python
data1 = data.fillna(data.mean()) # Imputando la media en los valores perdidos
```

```python
data1.isnull().sum() # Viendo si hay valores nulos:
# No hay
```
