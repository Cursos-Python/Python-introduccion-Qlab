```python
ndf_1 = ndf[['Departamento', 'nueva columna']]

ndf_2 = ndf2[['Departamento', 'Capital']]
```

```python
how = 'inner', 'left', 'right', 'outer'
```

```python
resultado_merge = pd.merge(ndf_1, ndf_2,
                    left_on = 'Departamento',
                    right_on = 'Departamento',
                    how = 'inner', indicator = True)
resultado_merge
```

```python
resultado_merge = pd.merge(ndf_1, ndf_2, on = 'Departamento', how = 'outer', indicator = True)
```
